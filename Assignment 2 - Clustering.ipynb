# Part 2: Performance Comparison

```python
import KMeans as km
import numpy as np
import matplotlib.pyplot as plt 

# Generating clusters using the make_blobs function
from sklearn.datasets import make_blobs
X, cluster_assignments = make_blobs(n_samples=700, centers=4, cluster_std=0.60, random_state=0)

print (cluster_assignments)
```

```python
# Plotting the clusters generated by make_blobs
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments, s=50, cmap='Pastel2')
plt.title('Figure 1: As generated from make_blobs')
plt.show()
```

```python
# My implementation of KMeans
kmeans = km.KMeans(k=4, max_iterations=100)
cluster_assignments_predicted, centroids = kmeans.fit(X)

print("My hypotheses:")
print()
print("Cluster assignments:", np.array(cluster_assignments_predicted))
print()
print("Centroids:", np.array(centroids))
```

```python
# Plotting the clusters generated by my implementation of KMeans
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments_predicted, s=50, cmap='Pastel2')
plt.title('Figure 2: As generated from my implementation of KMeans')
plt.show()
```

```python
# Implementing KMeans using scikit-learn
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, n_init=100)
kmeans.fit(X)
cluster_assignments_sklearn = kmeans.predict(X)

print("Scikit-learn's hypotheses:")
print()
print("Cluster assignments:", cluster_assignments_sklearn)
print()
print("Centroids:", kmeans.cluster_centers_)
```

```python
# Plotting the clusters generated by sklearn's KMeans
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments_sklearn, s=50, cmap='Pastel2')
plt.title('Figure 3: As generated from sklearn KMeans')
plt.show()
```

```python
# Metrics - Adjusted Rand Index (ARI) of my implementation of KMeans and that of scikit-learn's implementation
from sklearn.metrics import adjusted_rand_score

ari_predicted = adjusted_rand_score(cluster_assignments, cluster_assignments_predicted)
ari_sklearn = adjusted_rand_score(cluster_assignments, cluster_assignments_sklearn)
print(f"ARI for my implementation of KMeans: {ari_predicted}")
print(f"ARI for scikit-learn implementation: {ari_sklearn}")
```

## A - Chicago Taxi Data

```python
# Getting the data ready for implementing KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

df = pd.read_csv('Chicago_Taxi_Data.csv')
df

# We filter the DataFrame to only include the columns 'Pickup Centroid Latitude' and 'Pickup Centroid Longitude'
df = df.filter(['Pickup Centroid Latitude', 'Pickup Centroid Longitude'], axis=1)

# Drop rows with missing values
df = df.dropna()
df
```

```python
# Plotting pickup location data according to the pickup longitude and latitude
plt.xlabel('Pickup Centroid Longitude')
plt.ylabel('Pickup Centroid Latitude')
plt.title('Pickup Location - Chicago Taxi')
plt.scatter(df['Pickup Centroid Longitude'], df['Pickup Centroid Latitude'], s=6)
```

### Elbow Curve to determine the optimal number of clusters

```python
# Implementing the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
import pandas as pd
kmeans_inertia = pd.DataFrame(data=[], index=range(2,31), columns=['inertia'])
for n_clusters in range(2, 31):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10)
    kmeans.fit(df.iloc[:,:])
    kmeans_inertia.loc[n_clusters] = kmeans.inertia_
kmeans_inertia.plot(kind='line', y='inertia')
plt.xlabel('# of clusters')
plt.show()
```

The elbow seems to be between 6 and 8, so we can choose 7 as the optimal number of clusters.

### K-Means Clustering for Chicago Taxi Data

```python
# Implementing KMeans using scikit-learn and plotting the clusters
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=7, n_init=10)
kmeans.fit(df)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(df['Pickup Centroid Longitude'], df['Pickup Centroid Latitude'], c=labels, alpha=0.6, s=5, cmap="Set2")
plt.scatter(centroids[:,1], centroids[:,0], marker='x', c='black')
plt.xlabel('Pickup Centroid Longitude')
plt.ylabel('Pickup Centroid Latitude')
plt.title('Pickup Location of Chicago Taxi - KMeans Clusters with centroids')
plt.show()
```

### Silhouette Score

```python
# Mesuring the performance of KMeans using the Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_score(df, kmeans.labels_)
```

### Answers to questions based on Chicago Taxi Data

* **Reason for Choice of Clustering Algorithm(s):**
  - For this dataset, we use KMeans clustering algorithm because it is a popular and simple clustering algorithm that can handle large datasets efficiently and identify distinct groups. The dataset consists of geographic coordinates, which are continuous variables, and k-means is particularly effective in identifying clusters in continuous variable datasets and spatial data.

* **Pre-processing of Data :**
  - We filtered the dataframe columns to include only the pickup latitude and pickup longitude.
  - Rows with any missing values were dropped.
  - Plot the elbow curve to determine the optimal number of clusters, which in this case is 7.
    
* **Hyperparameter Settings:**
  - The hyperparameters include the number of clusters (k) based on the elbow curve. 

* **Output from the algorithm:**
  - The clusters have been plotted and seem evident that it mapped to the regions of Chicago. 

* **Metrics and Evaluation:**
  * The metrics used for this dataset Silhouette Score. A Score of around 0.68 is relatively high and close to 1. This implies that the clusters generated by the KMeans algorithm have well-defined and distinct boundaries, and the data points within each cluster are more similar to each other than to points in neighboring clusters. 
  * On comparing it with the 'Pickup Centroid Longitude vs. Pickup Centroid Latitude ' plotting for kmeans clustering, it can be seen that the kmeans clustering algorithm has divided the datapoints into different regions identical to that of the map of Chicago. ## B - Mopsi Data Subset - Teacking Individuals Routes around Finland

```python
# Getting the data ready for implementing DBSCAN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import sys

# Load the Finnish location data from Mopsi_Data_Finland.txt
data = pd.read_csv('Mopsi_Data_Finland.txt', sep="\t", header=None, names=["Latitude", "Longitude"])
data
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Normalizing the data
data = data/10000
data
```

```python
# Plotting the data to see the distribution of the points
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.3, s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Longitude vs Latitude')
plt.show
```

### DBSCAN Clustering for Mopsi Data Subset

```python
# Implementing DBSCAN using scikit-learn
from sklearn.cluster import DBSCAN
max_dist = 0.3
hoodsize = 20
cluster = DBSCAN(eps=max_dist, min_samples=hoodsize)
cluster.fit(data)
```

```python
cluster.labels_
```

```python
labels, count = np.unique(cluster.labels_, return_counts=True)
for label, count in zip(labels, count):
    print(label, " : ", count)
```

```python
# Plotting the clusters generated by DBSCAN
plt.scatter(data['Longitude'], data['Latitude'], c=cluster.labels_, alpha=0.4, s=8, cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Longitude vs Latitude')
plt.legend([x for x in cluster.labels_])
plt.show
```

### Silhouette Score

```python
# Mesuring the performance of DBSCAN using the silhouette score
from sklearn.metrics import silhouette_score
silhouette_score(data, cluster.labels_)
```

### Answers to questions based on Mopsi Data Subset


* **Reason for Choice of Clustering Algorithm(s):**
  - In the Mopsi dataset, where routes consist of diverse activities such as walking, cycling, skiing, and vehicle transport, clusters may not be well-defined spheres. DBSCAN's ability to detect clusters of varying shapes makes it suitable for such scenarios. DBSCAN automatically determines the number of clusters based on the density of the data points. It can detect clusters of different sizes, which is beneficial when dealing with routes that may have varying lengths and durations.

* **Pre-processing of Data :**
  - Checking for missing values.
  - Normalizing the data.

* **Hyperparameter Settings:**
  - Setting the esp and min_samples hyperparameters for DBSCAN.

* **Output from the algorithm:**
  - From the plotted graph, we can see that DBSCAN clustering gives us multiple clusters and giving optimal results for mapping the roadmaps in Finland. 

* **Metrics and Evaluation:**
  - The metric used for this dataset is the silhouette score and got a score of around 0.7 which is a good score and clustering result is quite good, with well-separated clusters and minimal overlap.
  - The clustering algorithm seems to find roads/routes and mapped each of them to a cluster which is exactly what we want to achieve with this dataset. Therefore, DBSCAN looks like the ideal clustering algorithm for this dataset.
